```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split 
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dragonheir/logistic-regression")

print("Path to dataset files:", path)
```

```python
weatherdf = pd.read_csv(path +"/Social_Network_Ads.csv")
```

```python
weatherdf
```

```python
weatherdf.drop(["User ID"],axis=1,inplace=True)
```

```python
weatherdf
```

```python
weatherdf["Gender"][weatherdf["Gender"] == "Male"] = 1 
weatherdf["Gender"][weatherdf["Gender"] == "Female"] = 0 
weatherdf["Gender"] = weatherdf.astype("int64")
```

```python
weatherdf.info()
```

```python
X = weatherdf.drop(["Purchased"],axis=1)
Y = weatherdf["Purchased"]
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.20,random_state=43)
```

```python
LR = LogisticRegression()
```

```python
LR.fit(X_train,Y_train)
```

```python
pred=LR.predict(X_test)
```

```python
Y_test
```

```python
from sklearn.metrics import classification_report
classification_report(Y_test, pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, pred)
```
